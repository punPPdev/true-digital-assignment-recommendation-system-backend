```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import mlflow
import mlflow.pyfunc
import warnings

# warnings.filterwarnings('ignore')

# ==========================================
# 1. THE CUSTOM PYFUNC MODEL
# ==========================================
class ContentBaseRecommenderModel(mlflow.pyfunc.PythonModel):
    """
    This custom MLflow model encapsulates the vectorizer, the pre-computed 
    catalog vectors, the catalog data, and the similarity logic.
    """
    def __init__(self, vectorizer, content_vector, content_data):
        self.vectorizer = vectorizer
        self.content_vector = content_vector
        self.content_data = content_data

    def predict(self, context, model_input):
        """
        MLflow requires this exact method signature.
        model_input is expected to be a pandas DataFrame sent from the API.
        """
        # Parse the input from the API
        # content_data = model_input.content_data.iloc[0]
        liked_content_str = model_input['liked_content'].iloc[0]
        top_k = model_input['top_k'].iloc[0]

        # 1. Vectorize the user's genre string
        user_vector = self.vectorizer.transform([liked_content_str])
        
        # 2. Compute Cosine Similarity inside the model
        similarities = cosine_similarity(user_vector, self.content_vector).flatten()
        
        # 3. Rank and Filter
        recommendations = self.content_data.copy()
        recommendations['similarity_score'] = similarities
        
        # # Exclude already liked titles
        # recommendations = temp_df #[~temp_df['title'].isin(liked_titles)]
        
        # Return the top K as a pandas DataFrame
        top_recs = recommendations.sort_values(by='similarity_score', ascending=False).head(top_k)
        return top_recs

# ==========================================
# 2. TRAINING AND LOGGING
# ==========================================
def train_and_pack_model():

    # 2. Train the Vectorizer
    vectorizer = TfidfVectorizer(stop_words='english')
    content_vector = vectorizer.fit_transform(content_txt)

    # 3. Instantiate our Custom PyFunc Model
    full_model = ContentBaseRecommenderModel(
        vectorizer=vectorizer,
        content_vector=content_vector,
        content_data=concat_tv_movie
    )

    # 4. Log to MLflow
    mlflow.set_experiment("PyFunc_Recommender_Experiment")
    with mlflow.start_run(run_name="Full_Pipeline_Model"):
        
        # Note: We use mlflow.pyfunc.log_model instead of mlflow.sklearn
        mlflow.pyfunc.log_model(
            artifact_path="full_recommender",
            python_model=full_model,
            registered_model_name="EndToEndGenreModel"
        )
        print("Successfully packed vectorizer, catalog data, and cosine similarity into a single MLflow model!")

if __name__ == "__main__":
    train_and_pack_model()
```
